```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Create dataframe with no duplicates in "SN" column
no_duplicates_df = purchase_data.drop_duplicates(subset="SN", keep="last")
#Find total number of unique players in newly created "SN" column
player_count = pd.DataFrame({"Total Players": [len(no_duplicates_df)]})
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Total Unique Items
unique_items = len(purchase_data["Item Name"].unique())
#Average Price
average_price = sum(purchase_data["Price"])/len(purchase_data["Price"])
#Number of Purchases
purchases = len(purchase_data["Purchase ID"])
#Total Revenue
total_revenue = sum(purchase_data["Price"])

# Purchasing Analysis dataframe
purchasing_analysis = pd.DataFrame([
    {"Number of Unique Items": unique_items,
    "Average Price": average_price,
    "Number of Purchases": purchases,
    "Total Revenue": total_revenue}
])
#Formatting
purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].map("${:.2f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map("${:.2f}".format)
purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#pull gender counts and total count of gender from no duplicates dataframe
gender_counts = no_duplicates_df["Gender"].value_counts()
gender_total = no_duplicates_df["Gender"].count()
#variables using gender count formulas
count_male = gender_counts["Male"]
percent_male = (count_male / gender_total)*100
count_female = gender_counts["Female"]
percent_female = (count_female / gender_total)*100
count_other = gender_counts["Other / Non-Disclosed"]
percent_other = (count_other / gender_total)*100

#create dataframe with gender demographics
gender_demographics = pd.DataFrame([
    {"Total Count": count_male, "Percentage of Players": percent_male},
    {"Total Count": count_female, "Percentage of Players": percent_female},
    {"Total Count": count_other, "Percentage of Players": percent_other}], index=["Male", "Female", "Other/Non-Disclosed"])
#Formatting
gender_demographics["Percentage of Players"] = gender_demographics["Percentage of Players"].map("{:.2f}%".format)
gender_demographics
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create Bins and Labels
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
ages = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#New dataframe for age analyses
no_duplicates_copy = no_duplicates_df.copy()
no_duplicates_copy["Age Ranges"] = pd.cut(no_duplicates_copy["Age"], bins, labels=ages)
no_duplicates_age = no_duplicates_copy.groupby("Age Ranges")
#Count and Formulas
age_dem_df = no_duplicates_age[["Age"]].count()
age_dem_df["Percentage of Players"] = (age_dem_df["Age"]/sum(age_dem_df["Age"]))*100
#Formatting
age_dem_df["Percentage of Players"] = age_dem_df["Percentage of Players"].map("{:.2f}%".format)

age_dem_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#New dataframe for purchasing analysis with full purchase list
purchase_data_copy = purchase_data.copy()
purchase_data_copy["Age Ranges"] = pd.cut(purchase_data_copy["Age"], bins, labels=ages)
purchase_data_pa = purchase_data_copy.groupby("Age Ranges")
#Count and Formulas
purchasing_analysis = purchase_data_pa[["Age"]].count()
purchasing_analysis["Average Purchase Price"] = purchase_data_pa[["Price"]].mean()
purchasing_analysis["Total Purchase Value"] = purchase_data_pa[["Price"]].sum()
purchasing_analysis["Avg Total Purchase per Person"] = purchasing_analysis["Total Purchase Value"]/age_dem_df["Age"]
#Formatting
purchasing_analysis["Average Purchase Price"] = purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis["Total Purchase Value"] = purchasing_analysis["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis["Avg Total Purchase per Person"] = purchasing_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
purchasing_analysis = purchasing_analysis.rename(columns = {"Age" : "Purchase Count"})

purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create groupby
pd_group = purchase_data.groupby("SN")
#Formulas
pd_df = pd.DataFrame(pd_group["SN"].count())
pd_df["Total Purchase Value"] = pd_group["Price"].sum()
pd_df["Average Purchase Price"] = pd_df["Total Purchase Value"]/pd_df["SN"]
# Rearrange and relabel columns
pd_df = pd_df.rename(columns={"SN" : "Purchase Count"})
pd_df = pd_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
#Formatting
pd_df = pd_df.sort_values(by="Total Purchase Value", ascending=False)
pd_df["Average Purchase Price"] = pd_df["Average Purchase Price"].map("${:.2f}".format)
pd_df["Total Purchase Value"] = pd_df["Total Purchase Value"].map("${:.2f}".format)

pd_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Created new dataframe with groupby
popular_items_group = purchase_data.groupby(["Item ID", "Item Name"])
pi_df = pd.DataFrame(popular_items_group["Item ID"].count())
pi_df = pi_df.rename(columns={"Item ID":"Purchase Count"})
#merge with original data to obtain price
popular_items = pd.merge(pi_df, purchase_data, on=["Item ID","Item Name"], how="left")
popular_items = popular_items[["Item ID","Item Name", "Purchase Count", "Price"]]
#drop duplicates and set new index
popular_items = popular_items.drop_duplicates(subset="Item Name", keep="first")
popular_items = popular_items.set_index(["Item ID", "Item Name"])
popular_items["Total Purchase Value"] = popular_items["Purchase Count"]*popular_items["Price"]
#Sort and format
popular_items_clean = popular_items.sort_values(by="Purchase Count", ascending=False)
popular_items_clean["Price"] = popular_items["Price"].map("${:.2f}".format)
popular_items_clean["Total Purchase Value"] = popular_items["Total Purchase Value"].map("${:.2f}".format)
popular_items_clean
```
